```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
import pandas as pd
from google.colab import files

print("Please upload your 'Salary' file now:")
uploaded = files.upload()

# Assuming the desired salary file is the first one uploaded.
# If you uploaded multiple files or the filename is different,
# please update `list(uploaded.keys())[0]` with the correct filename.
if uploaded:
    uploaded_filename = list(uploaded.keys())[0]
    print(f"Reading file: {uploaded_filename}")
    df = pd.read_csv(uploaded_filename)
    display(df.head())
else:
    print("No file was uploaded.")
```

```python
plt.scatter(df['YearsExperience'], df['Salary'], color='red')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs Experience')
plt.show()
```

```python
# Independent and Dependent variables
# Handle missing values by dropping rows where 'YearsExperience' is NaN
df_cleaned = df.dropna(subset=['YearsExperience']).copy()
X = df_cleaned[['YearsExperience']]
y = df_cleaned['Salary']

# Model creation
model = LinearRegression()
model.fit(X, y)
```

```python
predicted_salary = model.predict([[12]])

print("Predicted Salary for 12 years experience =", predicted_salary[0])
```

```python
plt.scatter(X, y, color='red')
plt.plot(X, model.predict(X), color='blue')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Regression Line')
plt.show()
```

```python
import pandas as pd
from google.colab import files

print("Please upload your 'hiring.csv' file now:")
uploaded_hiring = files.upload()

if uploaded_hiring:
    hiring_filename = list(uploaded_hiring.keys())[0]
    print(f"Reading file: {hiring_filename}")
    df_hiring = pd.read_csv(hiring_filename)
    display(df_hiring.head())
else:
    print("No file was uploaded.")
```

```python
# Fill missing experience with zero
df_hiring['experience'] = df_hiring['experience'].fillna("zero")

# Convert words to numbers
df_hiring['experience'] = df_hiring['experience'].replace({
    'zero':0,
    'one':1,
    'two':2,
    'three':3,
    'four':4,
    'five':5,
    'six':6,
    'seven':7,
    'eight':8,
    'nine':9,
    'ten':10,
    'eleven':11,
    'twelve':12
})

# Fill missing test score with median
median_test_score = math.floor(df_hiring['test_score(out of 10)'].median())
df_hiring['test_score(out of 10)'] = df_hiring['test_score(out of 10)'].fillna(median_test_score)

df_hiring
```

```python
model = LinearRegression()

X_hiring = df_hiring[['experience','test_score(out of 10)','interview_score(out of 10)']]
y_hiring = df_hiring['salary($)']

model.fit(X_hiring,y_hiring)
```

```python
model.predict([[2,9,6]])
```
